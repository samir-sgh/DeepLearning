```python
import pandas as pd

df = pd.read_csv("D:\Master enset\Deep Learning\devoir Deep Learning\Dataset of Diabetes .csv")
print(df.head())
```

```python
# Afficher des informations générales sur les données
print(df.info())

# Afficher des statistiques descriptives
print(df.describe())

# Vérifier s'il y a des valeurs manquantes
print(df.isnull().sum())
```

```python
print(df.columns)  # Liste des colonnes du dataset
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Affichage de la distribution de la variable cible
plt.figure(figsize=(8, 5))
sns.histplot(df["AGE"], bins=30, kde=True)
plt.title("Distribution de la variable cible (niveau de diabète)")
plt.show()

# Matrice de corrélation
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```
